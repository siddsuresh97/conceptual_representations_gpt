```python
import pandas as pd
import numpy as np
```

```python
trips = pd.read_csv('../../results/reptile_tool/davinci_triplet_feature_list_pruned.csv', index_col=1)
```

```python
trips
```

```python
items = np.unique(trips.to_numpy())
```

```python
item_ind_dict = dict(zip(items,range(items.shape[0])))
```

```python
def make_inds(x):
    y = [item_ind_dict[i] for i in x.values]
    return(y)
```

```python
trips2 = trips.apply(make_inds)
```

```python
trips2.to_csv('tim_triplets_2_inds.csv',index=None)
```

```python
for i,row in trips2.iterrows():
    print(f"- [{row[0]},{row[1]},{row[2]}]")
```

```python
for i in item_ind_dict.keys():
    print(f'"{i}",')
```
