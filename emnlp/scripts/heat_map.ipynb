```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

names = ['Human norms', 'Word2Vec', 'CLIP', 'GPT3(002)(unver)', 'GPT3(002)(ver)', 'Human triplet',
         'GPT3(002) triplet', 'GPT embedding', 'Human Pairwise', 'GPT3(003) Pairwise', 'FLAN-XXL Triplet',
         'FLAN-UL2 Triplet', 'FLAN-UL2 Pairwise', 'GPT3(002) Pairwise', 'FLAN-XXL(ver)']
task = [4, 1, 1, 4, 4, 3, 3, 1, 2, 2, 3, 3, 2, 2, 4]
name_to_task = dict(zip(names, task))
task_to_color = {1: 'red', 2: 'blue', 3: 'green', 4: 'orange'}  # Define your color mapping

# Load and process data for the first graph (remains the same)
corr = pd.read_csv('../results/dsm_no_cot_the_whole_correlation_table.csv', index_col=1, header=0)
corr = corr.iloc[:, 1:]
corr = corr.applymap(lambda x: x ** 2)
corr.index = names
corr.columns = names


human_f = corr['Human norms'].sort_values(ascending=False)
df = pd.DataFrame(index=human_f.index, columns=['correlation', 'task'])
df['correlation'] = human_f
df['task'] = df.index.map(name_to_task)

#remove indexes where the word 'Human' appears
df = df[~df.index.str.contains('Human')]
# remove the index is any of the values in the columns is NaN
df = df.dropna(axis=0, how='any')

# sort first by task. The order of the tasks is f, t, p, e
# then sort by correlation within each task
df = df.sort_values(by=['task', 'correlation'], ascending=False)
order = df.index

# Create the first plot
fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(10, 12))
axs[0].bar(df.index, df['correlation'], color=df['task'].map(task_to_color))
axs[0].set_title('Correlation of model representations with Human Feature Listing representations')
axs[0].set_xticklabels([])

# Load and process data for the second plot
human_t = corr['Human triplet'].sort_values(ascending=False)
df = pd.DataFrame(index=human_t.index, columns=['correlation', 'task'])
df['correlation'] = human_t
df['task'] = df.index.map(name_to_task)
#remove indexes where the word 'Human' appears
df = df[~df.index.str.contains('Human')]
# remove the index is any of the values in the columns is NaN
df = df.dropna(axis=0, how='any')

# use the same order as the first plot
df = df.reindex(order)

# Create the second plot below the first one
axs[1].bar(df.index, df['correlation'], color=df['task'].map(task_to_color))
axs[1].set_title('Correlation of model representations with Human Triplet representations')
axs[1].set_xticklabels([])

# Load and process data for the third plot
human_p = corr['Human Pairwise'].sort_values(ascending=False)
df = pd.DataFrame(index=human_p.index, columns=['correlation', 'task'])
df['correlation'] = human_p
df['task'] = df.index.map(name_to_task)

#remove indexes where the word 'Human' appears
df = df[~df.index.str.contains('Human')]
# remove the index is any of the values in the columns is NaN
df = df.dropna(axis=0, how='any')

# sort first by task. The order of the tasks is f, t, p, e
# then sort by correlation within each task
# df = df.sort_values(by=['task', 'correlation'], ascending=False)
df = df.reindex(order)

# Create the third plot below the second one
axs[2].bar(df.index, df['correlation'], color=df['task'].map(task_to_color))
axs[2].set_title('Correlation of model representations with Human Pairwise representations')
axs[2].set_xticklabels(df.index, rotation=90)

# Add the legend on the top chart
red_patch = mpatches.Patch(color='red', label='Feature Listing')
blue_patch = mpatches.Patch(color='blue', label='Triplet')
green_patch = mpatches.Patch(color='green', label='Pairwise')
orange_patch = mpatches.Patch(color='orange', label='Embedding')
axs[0].legend(handles=[red_patch, blue_patch, green_patch, orange_patch], loc='upper right', bbox_to_anchor=(1, 1), ncol=2)

# Adjust the layout and add space for labels
fig.tight_layout()
fig.subplots_adjust(bottom=0.2, left=0.1)

# Set a single y-label on the leftmost subplot
fig.text(0.02, 0.5, 'R^2', va='center', rotation='vertical', fontsize=14)

# Set a single x-label below the bottom subplot
fig.text(0.5, 0.05, 'Model representations', ha='center', fontsize=14)
#make the font such that it can be edited in illustrator
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42

#save the figure as a pdf with dpi 300
plt.savefig('../graphs/bar_graph.pdf', bbox_inches='tight', dpi=300)



```
